# Adult Census Income 보고서 보조 분석 노트북

이 노트북은 `Week-2-Data-Ops-EDA-and-Preprocessing.ipynb`를 바탕으로, 보고서에서 서술한 인사이트를 직접 재현하고 제출용 분석 코드로 남기기 위해 별도로 구성한 EDA 노트북이다.

핵심 목표는 다음과 같다.

- 데이터 개요와 클래스 불균형 확인
- `income` 라벨 정합성 문제 확인 및 정리
- `?` 기반 결측치 분석
- 수치형 변수 분포 및 IQR 이상치 비율 확인
- 범주형/수치형 변수와 `income`의 관계 분석
- 리스크 및 전처리 제안에 바로 연결되는 근거 표 생성

```python
import os
import sys
from pathlib import Path

extra_site = Path("..").resolve() / ".py38site"
if ".venv38" not in sys.executable and extra_site.exists():
    sys.path.insert(0, str(extra_site))

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno

from ucimlrepo import fetch_ucirepo

sns.set_theme(style="whitegrid", palette="Set2")
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.unicode_minus"] = False
```

```python
# 1. 데이터 로드
adult = fetch_ucirepo(id=2)
X = adult.data.features.copy()
y = adult.data.targets.copy()
df_raw = pd.concat([X, y], axis=1)

print("Raw shape:", df_raw.shape)
df_raw.head()
```

```python
# 2. 보고서용 기본 전처리
# - '?'를 NaN으로 변환
# - income 문자열의 마침표를 제거해 라벨 정합성 확보

df = df_raw.copy()

for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].replace("?", np.nan)

income_before = df["income"].value_counts(dropna=False).rename_axis("income_raw").reset_index(name="count")

df["income_clean"] = df["income"].str.replace(".", "", regex=False).str.strip()
df["income_binary"] = df["income_clean"].map({"<=50K": 0, ">50K": 1})

income_after = df["income_clean"].value_counts(dropna=False).rename_axis("income_clean").reset_index(name="count")

print("Unique raw income labels:", sorted(df_raw["income"].unique()))
print("Unique cleaned income labels:", sorted(df["income_clean"].dropna().unique()))
pd.concat({"before_cleaning": income_before, "after_cleaning": income_after}, axis=1)
```

```python
# 3. 데이터 개요 표
numeric_cols = [
    "age", "fnlwgt", "education-num", "capital-gain", "capital-loss", "hours-per-week"
]
categorical_cols = [
    "workclass", "education", "marital-status", "occupation",
    "relationship", "race", "sex", "native-country"
]

overview = pd.DataFrame({
    "항목": [
        "데이터 크기",
        "컬럼 구성",
        "수치형 변수",
        "범주형 변수",
        "타깃 변수"
    ],
    "내용": [
        f"{df.shape[0]:,} rows x {df_raw.shape[1]} columns",
        f"수치형 {len(numeric_cols)}개, 범주형 {len(categorical_cols)}개, 타깃 1개",
        ", ".join(numeric_cols),
        ", ".join(categorical_cols),
        "income"
    ]
})

class_ratio = (
    df["income_clean"]
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
    .rename("ratio_percent")
    .reset_index()
    .rename(columns={"index": "income_clean"})
)

display(overview)
display(class_ratio)
```

```python
# 4. 그림 1: income 타깃 분포
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.countplot(data=df, x="income", order=df["income"].value_counts().index, ax=axes[0])
axes[0].set_title("Raw income label distribution")
axes[0].tick_params(axis="x", rotation=20)

sns.countplot(data=df, x="income_clean", order=["<=50K", ">50K"], ax=axes[1])
axes[1].set_title("Cleaned income label distribution")
axes[1].tick_params(axis="x", rotation=0)

for ax in axes:
    for patch in ax.patches:
        height = int(patch.get_height())
        ax.annotate(height, (patch.get_x() + patch.get_width() / 2, height),
                    ha="center", va="bottom", fontsize=9, xytext=(0, 4), textcoords="offset points")

plt.tight_layout()
plt.show()
```

```python
# 5. 데이터 품질 분석: 결측치 분포
missing_summary = (
    df[categorical_cols + numeric_cols + ["income_clean"]]
    .isna()
    .sum()
    .sort_values(ascending=False)
    .rename("missing_count")
    .to_frame()
)
missing_summary["missing_ratio_percent"] = (missing_summary["missing_count"] / len(df) * 100).round(2)
missing_summary = missing_summary.query("missing_count > 0")

missing_summary
```

```python
# 6. 그림 2: 결측치 시각화
missing_cols = ["workclass", "occupation", "native-country"]

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

missing_ratio_plot = missing_summary.reset_index().rename(columns={"index": "column"})
sns.barplot(data=missing_ratio_plot, x="column", y="missing_ratio_percent", ax=axes[0])
axes[0].set_title("Missing ratio by column")
axes[0].set_ylabel("Missing ratio (%)")
axes[0].tick_params(axis="x", rotation=20)

plt.sca(axes[1])
msno.bar(df[missing_cols], color="#7fc97f")
axes[1].set_title("Observed values for missing-related columns")

plt.tight_layout()
plt.show()
```

```python
# 7. IQR 기준 이상치 비율 계산
def iqr_outlier_ratio(series: pd.Series) -> float:
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return ((series < lower) | (series > upper)).mean() * 100

outlier_table = pd.DataFrame({
    "feature": numeric_cols,
    "iqr_outlier_ratio_percent": [round(iqr_outlier_ratio(df[col].dropna()), 2) for col in numeric_cols],
    "median": [df[col].median() for col in numeric_cols],
    "q3": [df[col].quantile(0.75) for col in numeric_cols]
}).sort_values("iqr_outlier_ratio_percent", ascending=False)

outlier_table
```

```python
# 8. 그림 3: 주요 수치형 변수의 히스토그램 + 박스플롯
focus_numeric = ["age", "education-num", "hours-per-week", "capital-gain", "capital-loss"]

fig, axes = plt.subplots(len(focus_numeric), 2, figsize=(14, 4 * len(focus_numeric)))

for idx, col in enumerate(focus_numeric):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[idx, 0], color="#66c2a5")
    axes[idx, 0].set_title(f"Histogram: {col}")

    sns.boxplot(x=df[col], ax=axes[idx, 1], color="#fc8d62")
    axes[idx, 1].set_title(f"Boxplot: {col}")

plt.tight_layout()
plt.show()

df[focus_numeric + ["fnlwgt"]].describe().T
```

```python
# 9. 범주형 변수와 income 관계 분석
selected_cat_cols = ["education", "marital-status", "occupation", "relationship", "sex"]

cat_target_summary = []
for col in selected_cat_cols:
    tmp = (
        df.groupby(col, dropna=False)["income_binary"]
        .agg(sample_count="count", high_income_rate="mean")
        .reset_index()
    )
    tmp["high_income_rate"] = (tmp["high_income_rate"] * 100).round(2)
    tmp["feature"] = col
    cat_target_summary.append(tmp)

cat_target_summary = pd.concat(cat_target_summary, ignore_index=True)
cat_target_summary.head()
```

```python
# 10. 그림 4: 주요 범주형 변수별 P(>50K) 비교
fig, axes = plt.subplots(3, 2, figsize=(16, 18))
axes = axes.flatten()

plot_order_config = {
    "education": 12,
    "marital-status": 8,
    "occupation": 12,
    "relationship": 8,
    "sex": 8,
}

for idx, col in enumerate(selected_cat_cols):
    tmp = (
        df.groupby(col, dropna=False)["income_binary"]
        .agg(sample_count="count", high_income_rate="mean")
        .sort_values(["high_income_rate", "sample_count"], ascending=[False, False])
        .head(plot_order_config[col])
        .reset_index()
    )
    tmp["high_income_rate"] = tmp["high_income_rate"] * 100
    sns.barplot(data=tmp, y=col, x="high_income_rate", ax=axes[idx], color="#8da0cb")
    axes[idx].set_title(f"P(>50K) by {col}")
    axes[idx].set_xlabel("High income rate (%)")
    axes[idx].set_ylabel(col)

axes[-1].axis("off")
plt.tight_layout()
plt.show()
```

```python
# 11. 그림 5: 주요 수치형 변수와 income의 관계
target_numeric = ["age", "education-num", "hours-per-week", "capital-gain"]

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for idx, col in enumerate(target_numeric):
    sns.boxplot(data=df, x="income_clean", y=col, order=["<=50K", ">50K"], ax=axes[idx], palette="Set3")
    axes[idx].set_title(f"{col} by income")

plt.tight_layout()
plt.show()

numeric_target_summary = (
    df.groupby("income_clean")[target_numeric]
    .agg(["median", "mean"])
    .round(2)
)
numeric_target_summary
```

```python
# 12. 그림 6: 수치형 변수 상관관계
corr_cols = numeric_cols + ["income_binary"]
corr_matrix = df[corr_cols].corr(numeric_only=True)

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="RdBu_r", center=0, fmt=".2f")
plt.title("Correlation heatmap for numeric features")
plt.tight_layout()
plt.show()

corr_with_target = corr_matrix["income_binary"].sort_values(ascending=False).rename("corr_with_income_binary")
corr_with_target.to_frame()
```

```python
# 13. 중복 정보 및 희귀 범주 점검
education_alignment = (
    df.groupby(["education", "education-num"]).size().reset_index(name="count")
    .sort_values(["education-num", "count"], ascending=[True, False])
)

rare_country = (
    df["native-country"].value_counts(dropna=False)
    .rename_axis("native-country")
    .reset_index(name="count")
)
rare_country["ratio_percent"] = (rare_country["count"] / len(df) * 100).round(3)

print("Education / education-num 조합 예시")
display(education_alignment.head(15))

print("native-country 희귀 범주 점검")
display(rare_country.tail(15))
```

```python
# 14. 보고서 본문에 바로 옮길 수 있는 리스크/권장 대응 표
risk_table = pd.DataFrame([
    {
        "이슈": "income 라벨 불일치",
        "수준": "치명적",
        "근거": "<=50K, <=50K., >50K, >50K. 네 문자열이 혼재",
        "권장 대응": "마침표 제거 후 이진 라벨 재생성"
    },
    {
        "이슈": "occupation / workclass 결측",
        "수준": "중간",
        "근거": f"occupation {missing_summary.loc['occupation', 'missing_ratio_percent']:.2f}%, workclass {missing_summary.loc['workclass', 'missing_ratio_percent']:.2f}%",
        "권장 대응": "단순 삭제보다 Missing/Unknown 범주 보존"
    },
    {
        "이슈": "native-country 결측",
        "수준": "중간",
        "근거": f"native-country {missing_summary.loc['native-country', 'missing_ratio_percent']:.2f}%",
        "권장 대응": "결측 보존 + rare grouping 고려"
    },
    {
        "이슈": "hours-per-week IQR 이상치",
        "수준": "주의",
        "근거": f"IQR 이상치 비율 {outlier_table.set_index('feature').loc['hours-per-week', 'iqr_outlier_ratio_percent']:.2f}%",
        "권장 대응": "기계적 삭제 대신 의미 기반 해석"
    },
    {
        "이슈": "capital-gain / capital-loss long-tail",
        "수준": "주의",
        "근거": (
            f"IQR 이상치 비율 {outlier_table.set_index('feature').loc['capital-gain', 'iqr_outlier_ratio_percent']:.2f}% / "
            f"{outlier_table.set_index('feature').loc['capital-loss', 'iqr_outlier_ratio_percent']:.2f}%"
        ),
        "권장 대응": "로그 변환 또는 non-zero flag 추가 검토"
    }
])

risk_table
```

```python
# 15. 전처리 전략 제안 표
preprocessing_plan = pd.DataFrame([
    ["타깃 정리", "income의 마침표 제거 후 <=50K / >50K로 통일, binary target 생성"],
    ["결측 처리", "'?'를 NaN으로 바꾼 뒤 workclass/occupation/native-country는 Missing 또는 Unknown 범주 보존 우선 검토"],
    ["범주형 인코딩", "sex, race, workclass, marital-status, relationship는 one-hot encoding, native-country는 rare grouping 후 인코딩"],
    ["수치형 처리", "capital-gain, capital-loss는 log1p 또는 non-zero flag 추가 고려"],
    ["이상치 처리", "IQR 기준 기계적 삭제 금지, 의미 해석 우선"],
    ["중복 변수 점검", "education과 education-num은 동시 사용 여부를 실험적으로 비교"],
    ["평가 전략", "accuracy 외 precision, recall, F1, ROC-AUC 함께 확인"],
], columns=["단계", "제안"])

preprocessing_plan
```
